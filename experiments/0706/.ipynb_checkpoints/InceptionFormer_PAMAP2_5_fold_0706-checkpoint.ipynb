```python
#!/usr/bin/env python
# coding: utf-8

# PAMAP2 MKI_Rotation Model - 5-Fold Cross Validation
# Complete evaluation with detailed metrics and visualizations

import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import glob
import time
import gc

# For reproducibility with CUDA optimization
import random
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = False  # Set to False for better performance
        torch.backends.cudnn.benchmark = True  # Enable cuDNN benchmark for consistent input size

# ---------------- Rotation Transformation Functions ----------------
def create_rotation_matrix_z(angle_degrees):
    """Create rotation matrix for Z-axis rotation"""
    angle_rad = np.radians(angle_degrees)
    cos_a = np.cos(angle_rad)
    sin_a = np.sin(angle_rad)

    rotation_matrix = np.array([
        [cos_a, -sin_a, 0],
        [sin_a,  cos_a, 0],
        [0,      0,     1]
    ], dtype=np.float32)

    return rotation_matrix

def apply_rotation_to_3axis(data, rotation_matrix):
    """Apply rotation to 3-axis sensor data"""
    return data @ rotation_matrix.T

def apply_rotation_to_pamap2_data(sensor_data, rotation_angles=[0, 90, 180, 270]):
    """Apply multiple rotations to PAMAP2 sensor data"""
    rotated_data_list = []
    rotation_labels = []

    # Transpose to (seq_len, num_channels) for easier processing
    data_transposed = sensor_data.transpose(1, 0)

    for i, angle in enumerate(rotation_angles):
        rotation_matrix = create_rotation_matrix_z(angle)
        rotated_data = data_transposed.copy()

        # Apply rotation to Hand IMU data (channels 0-15: 4 sensors x 4 values - temp, acc_x, acc_y, acc_z, gyro_x, gyro_y, gyro_z, mag_x, mag_y, mag_z)
        # Hand accelerometer (channels 1-3)
        if sensor_data.shape[0] >= 4:
            hand_acc = data_transposed[:, 1:4]  # channels 1,2,3
            rotated_hand_acc = apply_rotation_to_3axis(hand_acc, rotation_matrix)
            rotated_data[:, 1:4] = rotated_hand_acc

        # Hand gyroscope (channels 4-6)
        if sensor_data.shape[0] >= 7:
            hand_gyro = data_transposed[:, 4:7]  # channels 4,5,6
            rotated_hand_gyro = apply_rotation_to_3axis(hand_gyro, rotation_matrix)
            rotated_data[:, 4:7] = rotated_hand_gyro

        # Hand magnetometer (channels 7-9)
        if sensor_data.shape[0] >= 10:
            hand_mag = data_transposed[:, 7:10]  # channels 7,8,9
            rotated_hand_mag = apply_rotation_to_3axis(hand_mag, rotation_matrix)
            rotated_data[:, 7:10] = rotated_hand_mag

        # Chest IMU data (channels 17-26)
        # Chest accelerometer (channels 18-20)
        if sensor_data.shape[0] >= 21:
            chest_acc = data_transposed[:, 18:21]
            rotated_chest_acc = apply_rotation_to_3axis(chest_acc, rotation_matrix)
            rotated_data[:, 18:21] = rotated_chest_acc

        # Chest gyroscope (channels 21-23)
        if sensor_data.shape[0] >= 24:
            chest_gyro = data_transposed[:, 21:24]
            rotated_chest_gyro = apply_rotation_to_3axis(chest_gyro, rotation_matrix)
            rotated_data[:, 21:24] = rotated_chest_gyro

        # Chest magnetometer (channels 24-26)
        if sensor_data.shape[0] >= 27:
            chest_mag = data_transposed[:, 24:27]
            rotated_chest_mag = apply_rotation_to_3axis(chest_mag, rotation_matrix)
            rotated_data[:, 24:27] = rotated_chest_mag

        # Ankle IMU data (channels 34-43)
        # Ankle accelerometer (channels 35-37)
        if sensor_data.shape[0] >= 38:
            ankle_acc = data_transposed[:, 35:38]
            rotated_ankle_acc = apply_rotation_to_3axis(ankle_acc, rotation_matrix)
            rotated_data[:, 35:38] = rotated_ankle_acc

        # Ankle gyroscope (channels 38-40)
        if sensor_data.shape[0] >= 41:
            ankle_gyro = data_transposed[:, 38:41]
            rotated_ankle_gyro = apply_rotation_to_3axis(ankle_gyro, rotation_matrix)
            rotated_data[:, 38:41] = rotated_ankle_gyro

        # Ankle magnetometer (channels 41-43)
        if sensor_data.shape[0] >= 44:
            ankle_mag = data_transposed[:, 41:44]
            rotated_ankle_mag = apply_rotation_to_3axis(ankle_mag, rotation_matrix)
            rotated_data[:, 41:44] = rotated_ankle_mag

        # Transpose back to (num_channels, seq_len)
        rotated_data_final = rotated_data.transpose(1, 0)

        rotated_data_list.append(rotated_data_final)
        rotation_labels.append(i)

    return rotated_data_list, rotation_labels

# ---------------- Basic Building Blocks ----------------
class DepthwiseSeparableConv(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size, padding, bias=False):
        super().__init__()
        self.depthwise = nn.Conv1d(in_channels, in_channels, kernel_size=kernel_size,
                                        padding=padding, groups=in_channels, bias=bias)
        self.pointwise = nn.Conv1d(in_channels, out_channels, kernel_size=1, bias=bias)
        self.bn = nn.BatchNorm1d(out_channels)
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        x = self.depthwise(x)
        x = self.pointwise(x)
        x = self.bn(x)
        return self.relu(x)

class MultiKernalBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.skip_proj = nn.Conv1d(in_channels, out_channels * 4, kernel_size=1, bias=False)
        self.bn_skip = nn.BatchNorm1d(out_channels * 4)
        self.branch7 = DepthwiseSeparableConv(in_channels, out_channels, kernel_size=7, padding=3)
        self.branch5 = DepthwiseSeparableConv(in_channels, out_channels, kernel_size=5, padding=2)
        self.branch3 = DepthwiseSeparableConv(in_channels, out_channels, kernel_size=3, padding=1)
        self.branch1 = DepthwiseSeparableConv(in_channels, out_channels, kernel_size=1, padding=0)

    def forward(self, x):
        identity = self.bn_skip(self.skip_proj(x))
        x7 = self.branch7(x)
        x5 = self.branch5(x)
        x3 = self.branch3(x)
        x1 = self.branch1(x)
        out = torch.cat([x1, x3, x5, x7], dim=1)
        out = out + identity
        return F.relu(out)

class InceptionBlock(nn.Module):
    def __init__(self, in_channels, out_channels_per_branch):
        super().__init__()
        total_out_channels = out_channels_per_branch * 4
        self.skip_proj = nn.Conv1d(in_channels, total_out_channels, kernel_size=1, bias=False)
        self.bn_skip = nn.BatchNorm1d(total_out_channels)
        self.branch1 = DepthwiseSeparableConv(in_channels, out_channels_per_branch, kernel_size=1, padding=0)
        self.branch3 = nn.Sequential(
            DepthwiseSeparableConv(in_channels, out_channels_per_branch, kernel_size=1, padding=0),
            DepthwiseSeparableConv(out_channels_per_branch, out_channels_per_branch, kernel_size=3, padding=1)
        )
        self.branch5 = nn.Sequential(
            DepthwiseSeparableConv(in_channels, out_channels_per_branch, kernel_size=1, padding=0),
            DepthwiseSeparableConv(out_channels_per_branch, out_channels_per_branch, kernel_size=5, padding=2)
        )
        self.pool = nn.Sequential(
            nn.MaxPool1d(kernel_size=3, stride=1, padding=1),
            DepthwiseSeparableConv(in_channels, out_channels_per_branch, kernel_size=1, padding=0)
        )

    def forward(self, x):
        identity = self.bn_skip(self.skip_proj(x))
        b1 = self.branch1(x)
        b3 = self.branch3(x)
        b5 = self.branch5(x)
        bp = self.pool(x)
        out = torch.cat([b1, b3, b5, bp], dim=1)
        out = out + identity
        return F.relu(out)

# ---------------- MKI_Rotation Model ----------------
class Model_MKI_Rotation(nn.Module):
    """Model: stem->MultiKernel->Inception->classifier (Rotation)"""
    def __init__(self, input_channels=48, feature_dim=64, num_classes=18, dropout=0.1):
        super().__init__()
        self.name = "MKI_Rotation"
        # Stem
        self.stem = nn.Sequential(
            nn.Conv1d(input_channels, feature_dim, kernel_size=7, padding=3, bias=False),
            nn.BatchNorm1d(feature_dim),
            nn.ReLU(inplace=True),
            nn.MaxPool1d(kernel_size=3, stride=2, padding=1)
        )
        # MultiKernel
        self.multikernel = MultiKernalBlock(feature_dim, feature_dim)
        # Inception
        self.inception = InceptionBlock(feature_dim * 4, feature_dim)
        # Global Average Pooling
        self.gap = nn.AdaptiveAvgPool1d(1)
        self.flatten = nn.Flatten()

        # Main classifier
        self.main_classifier = nn.Sequential(
            nn.LayerNorm(feature_dim * 4),
            nn.Linear(feature_dim * 4, 128),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(128, num_classes)
        )

        # Rotation classifier (4-class: 0, 90, 180, 270)
        self.rotation_classifier = nn.Sequential(
            nn.LayerNorm(feature_dim * 4),
            nn.Linear(feature_dim * 4, 64),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(64, 4)
        )

    def forward(self, x):
        x = self.stem(x)
        x = self.multikernel(x)
        x = self.inception(x)
        x = self.gap(x)
        features = self.flatten(x)

        main_output = self.main_classifier(features)
        rotation_output = self.rotation_classifier(features)

        return main_output, rotation_output, features

# ---------------- PAMAP2 Data Loading ----------------
def load_and_window_pamap2(dataset_dir, window_size=100, step=50, exclude_labels=[0]):
    """Load and window PAMAP2 dataset with all IMU sensors (Hand, Chest, Ankle)"""
    all_windows, all_labels = [], []
    original_labels_collected = []
    label_encoder = LabelEncoder()

    # Find Protocol and Optional files
    file_paths = sorted(glob.glob(os.path.join(dataset_dir, 'Protocol', 'subject*.dat')))
    file_paths += sorted(glob.glob(os.path.join(dataset_dir, 'Optional', 'subject*.dat')))

    if not file_paths:
        print(f"Error: No 'subject*.dat' files found in directory: {dataset_dir}/Protocol or {dataset_dir}/Optional.")
        return None, None, None, None

    print(f"Found {len(file_paths)} subject files. Processing...")

    for file_path in file_paths:
        subject_id = os.path.basename(file_path).split('.')[0]
        print(f"  Processing {subject_id}...")
        try:
            df = pd.read_csv(file_path, sep='\s+', header=None, na_values='NaN')
            print(f"    {subject_id}: Loaded dataframe with shape {df.shape}")

            # Drop rows with any NaN values
            df_cleaned = df.dropna()
            print(f"    {subject_id}: Shape after dropping NaNs: {df_cleaned.shape}")

            if df_cleaned.empty:
                print(f"    Skipping {subject_id}: No data remaining after cleaning NaN values.")
                continue

            # Extract labels and sensor features
            labels = df_cleaned.iloc[:, 1].values.astype(int)  # Activity label is in column 1

            # Select all IMU sensor columns (Hand + Chest + Ankle)
            # Hand IMU: columns 4-19 (16 channels)
            # Chest IMU: columns 21-36 (16 channels)
            # Ankle IMU: columns 38-53 (16 channels)
            hand_cols = list(range(4, 20))      # 4-19
            chest_cols = list(range(21, 37))    # 21-36
            ankle_cols = list(range(38, 54))    # 38-53
            all_sensor_cols = hand_cols + chest_cols + ankle_cols  # Total 48 channels

            # Check if we have enough columns
            if df_cleaned.shape[1] < max(all_sensor_cols) + 1:
                print(f"    Skipping {subject_id}: Not enough columns.")
                continue

            features = df_cleaned.iloc[:, all_sensor_cols].values.astype(np.float32)

            # Filter out excluded labels
            valid_indices = np.where(~np.isin(labels, exclude_labels))[0]
            if len(valid_indices) == 0:
                print(f"    Skipping {subject_id}: No valid activities.")
                continue

            features = features[valid_indices, :]
            labels = labels[valid_indices]
            print(f"    {subject_id}: {len(features)} valid samples after filtering.")
            original_labels_collected.extend(labels)

            if len(features) < window_size:
                print(f"    Skipping {subject_id}: Not enough data for one window.")
                continue

            # Create sliding windows
            subject_windows, subject_window_labels = [], []
            start = 0
            while start + window_size <= len(features):
                window_data = features[start : start + window_size, :]
                window_labels_raw = labels[start : start + window_size]

                if len(window_labels_raw) > 0:
                    most_common_label = Counter(window_labels_raw).most_common(1)[0][0]
                else:
                    start += step
                    continue

                subject_windows.append(window_data)
                subject_window_labels.append(most_common_label)
                start += step

            if subject_windows:
                all_windows.extend(subject_windows)
                all_labels.extend(subject_window_labels)
                print(f"    {subject_id}: Created {len(subject_windows)} windows.")

        except Exception as e:
            print(f"    Error processing {subject_id}: {e}")

    if not all_windows:
        print("Error: No windows were created.")
        return None, None, None, None

    # Encode labels
    label_encoder.fit(sorted(set(original_labels_collected)))
    y_encoded = label_encoder.transform(all_labels)
    class_names = list(label_encoder.classes_)
    num_classes_actual = len(np.unique(y_encoded))

    # Convert to numpy and transpose for Conv1D: (N, L, C) -> (N, C, L)
    X_windowed = np.array(all_windows, dtype=np.float32)
    X_windowed = np.transpose(X_windowed, (0, 2, 1))

    print(f"\nWindowing complete. Total windows: {X_windowed.shape[0]}")
    print(f"  Windowed data shape (N, C, L): {X_windowed.shape}")
    print(f"  Number of classes: {num_classes_actual}")

    # PAMAP2 18 activity names
    pamap2_activity_names = [
        'lying', 'sitting', 'standing', 'walking', 'running', 'cycling',
        'Nordic walking', 'watching TV', 'computer work', 'car driving',
        'ascending stairs', 'descending stairs', 'vacuum cleaning', 'ironing',
        'folding laundry', 'house cleaning', 'playing soccer', 'rope jumping'
    ]

    if len(pamap2_activity_names) != num_classes_actual:
        print(f"Warning: Activity names mismatch. Using found class names.")
        report_class_names = class_names
    else:
        report_class_names = pamap2_activity_names

    return X_windowed, y_encoded, report_class_names, num_classes_actual

# ---------------- Rotation Dataset ----------------
class RotationDataset(Dataset):
    def __init__(self, data, labels, label_ratio=1.0):
        self.original_data = torch.tensor(data, dtype=torch.float32)
        self.original_labels = torch.tensor(labels, dtype=torch.long)

        # Generate rotation data
        self.rotation_angles = [0, 90, 180, 270]
        self.all_data = []
        self.all_rotation_labels = []
        self.all_main_labels = []
        self.all_has_main_labels = []

        # Semi-supervised setting
        self.num_samples = len(data)
        self.num_labeled = int(self.num_samples * label_ratio)
        self.labeled_indices = set(range(self.num_labeled))

        for sample_idx in range(self.num_samples):
            original_sample = data[sample_idx]
            rotated_samples, rotation_labels = apply_rotation_to_pamap2_data(original_sample, self.rotation_angles)

            for rot_sample, rot_label in zip(rotated_samples, rotation_labels):
                self.all_data.append(torch.tensor(rot_sample, dtype=torch.float32))
                self.all_rotation_labels.append(rot_label)

                # Only original rotation (0) gets main labels for labeled samples
                if sample_idx in self.labeled_indices and rot_label == 0:
                    self.all_main_labels.append(labels[sample_idx])
                    self.all_has_main_labels.append(True)
                else:
                    self.all_main_labels.append(-1)
                    self.all_has_main_labels.append(False)

        self.all_rotation_labels = torch.tensor(self.all_rotation_labels, dtype=torch.long)
        self.all_main_labels = torch.tensor(self.all_main_labels, dtype=torch.long)
        self.all_has_main_labels = torch.tensor(self.all_has_main_labels, dtype=torch.bool)

    def __len__(self):
        return len(self.all_data)

    def __getitem__(self, idx):
        return (self.all_data[idx], self.all_main_labels[idx],
                self.all_rotation_labels[idx], self.all_has_main_labels[idx])

# ---------------- Training Function with CUDA Optimization ----------------
def train_self_supervised_model(model, train_loader, val_loader, device, epochs=30, patience=5, alpha=1.0, lr=0.001):
    model = model.to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=0.05)  # Increased weight decay
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', factor=0.7, patience=3)  # Removed verbose parameter

    main_criterion = nn.CrossEntropyLoss(ignore_index=-1, reduction='mean')
    pretext_criterion = nn.CrossEntropyLoss(reduction='mean')  # 4-class rotation

    # Mixed precision training for GPU optimization
    use_amp = torch.cuda.is_available()
    scaler = torch.amp.GradScaler('cuda') if use_amp else None

    best_val_loss = float('inf')
    epochs_no_improve = 0
    best_model_state = None

    for epoch in range(epochs):
        # Training
        model.train()
        running_total_loss = 0.0
        correct_main = 0
        total_main = 0

        for batch_data in train_loader:
            ts_data, main_labels, pretext_labels, has_main_labels = batch_data
            ts_data = ts_data.to(device, non_blocking=True)
            main_labels = main_labels.to(device, non_blocking=True)
            pretext_labels = pretext_labels.to(device, non_blocking=True)

            optimizer.zero_grad()

            # Mixed precision forward pass
            if use_amp:
                with torch.amp.autocast('cuda'):
                    main_outputs, pretext_outputs, _ = model(ts_data)
                    pretext_loss = pretext_criterion(pretext_outputs, pretext_labels)
                    labeled_mask = (main_labels != -1)
                    if labeled_mask.sum() > 0:
                        main_loss = main_criterion(main_outputs, main_labels)
                    else:
                        main_loss = torch.tensor(0.0, device=device, requires_grad=True)
                    total_loss = main_loss + alpha * pretext_loss

                # Mixed precision backward pass
                scaler.scale(total_loss).backward()
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.5)  # Stricter gradient clipping
                scaler.step(optimizer)
                scaler.update()
            else:
                main_outputs, pretext_outputs, _ = model(ts_data)
                pretext_loss = pretext_criterion(pretext_outputs, pretext_labels)
                labeled_mask = (main_labels != -1)
                if labeled_mask.sum() > 0:
                    main_loss = main_criterion(main_outputs, main_labels)
                else:
                    main_loss = torch.tensor(0.0, device=device, requires_grad=True)
                total_loss = main_loss + alpha * pretext_loss
                total_loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.5)  # Stricter gradient clipping
                optimizer.step()

            running_total_loss += total_loss.item() * ts_data.size(0)

            if labeled_mask.sum() > 0:
                with torch.no_grad():
                    _, predicted = torch.max(main_outputs[labeled_mask], 1)
                    total_main += labeled_mask.sum().item()
                    correct_main += (predicted == main_labels[labeled_mask]).sum().item()

        epoch_train_loss = running_total_loss / len(train_loader.dataset)
        epoch_train_acc = correct_main / total_main if total_main > 0 else 0.0

        # Validation
        model.eval()
        running_val_loss = 0.0
        correct_val = 0
        total_val = 0

        with torch.no_grad():
            for batch_data in val_loader:
                ts_data, main_labels, pretext_labels, has_main_labels = batch_data
                ts_data = ts_data.to(device, non_blocking=True)
                main_labels = main_labels.to(device, non_blocking=True)
                pretext_labels = pretext_labels.to(device, non_blocking=True)

                if use_amp:
                    with torch.amp.autocast('cuda'):
                        main_outputs, pretext_outputs, _ = model(ts_data)
                        pretext_loss = pretext_criterion(pretext_outputs, pretext_labels)
                        labeled_mask = (main_labels != -1)
                        if labeled_mask.sum() > 0:
                            main_loss = main_criterion(main_outputs, main_labels)
                        else:
                            main_loss = torch.tensor(0.0, device=device)
                        total_loss = main_loss + alpha * pretext_loss
                else:
                    main_outputs, pretext_outputs, _ = model(ts_data)
                    pretext_loss = pretext_criterion(pretext_outputs, pretext_labels)
                    labeled_mask = (main_labels != -1)
                    if labeled_mask.sum() > 0:
                        main_loss = main_criterion(main_outputs, main_labels)
                    else:
                        main_loss = torch.tensor(0.0, device=device)
                    total_loss = main_loss + alpha * pretext_loss

                running_val_loss += total_loss.item() * ts_data.size(0)

                if labeled_mask.sum() > 0:
                    _, predicted = torch.max(main_outputs[labeled_mask], 1)
                    total_val += labeled_mask.sum().item()
                    correct_val += (predicted == main_labels[labeled_mask]).sum().item()

        epoch_val_loss = running_val_loss / len(val_loader.dataset)
        epoch_val_acc = correct_val / total_val if total_val > 0 else 0.0

        if epoch % 5 == 0 or epoch < 10:  # More frequent early logging
            gpu_memory = torch.cuda.memory_allocated() / 1024**3 if torch.cuda.is_available() else 0
            print(f"  Epoch {epoch+1}/{epochs} - Train Loss: {epoch_train_loss:.4f}, Train Acc: {epoch_train_acc:.4f}, Val Loss: {epoch_val_loss:.4f}, Val Acc: {epoch_val_acc:.4f}, GPU Mem: {gpu_memory:.2f}GB")

        scheduler.step(epoch_val_loss)

        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss
            epochs_no_improve = 0
            best_model_state = model.state_dict().copy()
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print(f"  Early stopping after {epoch+1} epochs.")
                break

    if best_model_state:
        model.load_state_dict(best_model_state)

    return model

# ---------------- Evaluation Function with CUDA Optimization ----------------
def evaluate_model(model, test_loader, device, class_names, fold_idx):
    model.eval()
    all_preds, all_labels = [], []

    with torch.no_grad():
        for batch_data in test_loader:
            ts_data, main_labels, pretext_labels, has_main_labels = batch_data
            ts_data = ts_data.to(device, non_blocking=True)
            main_labels = main_labels.to(device, non_blocking=True)

            # Use mixed precision for inference if available
            if torch.cuda.is_available():
                with torch.amp.autocast('cuda'):
                    main_outputs, _, _ = model(ts_data)
            else:
                main_outputs, _, _ = model(ts_data)

            labeled_mask = (main_labels != -1)
            if labeled_mask.sum() > 0:
                _, predicted = torch.max(main_outputs[labeled_mask], 1)
                all_preds.extend(predicted.cpu().numpy())
                all_labels.extend(main_labels[labeled_mask].cpu().numpy())

    # Calculate metrics
    accuracy = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='macro', zero_division=0)
    precision = precision_score(all_labels, all_preds, average='macro', zero_division=0)
    recall = recall_score(all_labels, all_preds, average='macro', zero_division=0)

    print(f"Fold {fold_idx+1} Results:")
    print(f"  Accuracy: {accuracy:.4f}")
    print(f"  F1 Score: {f1:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall: {recall:.4f}")

    return {
        'accuracy': accuracy,
        'f1': f1,
        'precision': precision,
        'recall': recall,
        'predictions': all_preds,
        'labels': all_labels
    }

# ---------------- Visualization Functions ----------------
def plot_confusion_matrix(y_true, y_pred, class_names, fold_idx):
    """Plot normalized confusion matrix for each fold"""
    cm = confusion_matrix(y_true, y_pred)
    cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100

    # Calculate dynamic figure size based on number of classes
    n_classes = len(class_names)
    fig_size = max(12, n_classes * 0.8)

    plt.figure(figsize=(fig_size, fig_size))
    sns.heatmap(
        cm_norm,
        annot=True,
        fmt=".1f",
        cmap="Blues",
        xticklabels=class_names,
        yticklabels=class_names,
        cbar_kws={'label': 'Percentage (%)'},
        square=True,  # Make cells square
        linewidths=0.5  # Add lines between cells
    )
    plt.title(f'MKI_Rotation Fold {fold_idx+1} - Confusion Matrix (%)',
              fontsize=16, weight='bold', pad=20)
    plt.xlabel('Predicted Label', fontsize=12, labelpad=10)
    plt.ylabel('True Label', fontsize=12, labelpad=10)

    # Adjust label rotation and alignment
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(rotation=0, fontsize=10)

    # Use constrained_layout for better spacing
    plt.tight_layout()

    # Adjust margins to prevent cutting
    plt.subplots_adjust(bottom=0.15, left=0.15, right=0.95, top=0.92)

    plt.show()

def plot_cv_results(cv_results):
    """Plot cross-validation results summary"""
    metrics = ['accuracy', 'f1', 'precision', 'recall']

    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    axes = axes.flatten()

    for i, metric in enumerate(metrics):
        fold_values = [cv_results[f'fold_{j+1}'][metric] for j in range(5)]
        mean_val = np.mean(fold_values)
        std_val = np.std(fold_values)

        x_pos = range(1, 6)
        bars = axes[i].bar(x_pos, fold_values, color='#3498db', alpha=0.7)
        axes[i].axhline(y=mean_val, color='red', linestyle='--', linewidth=2,
                       label=f'Mean: {mean_val:.4f}  {std_val:.4f}')

        axes[i].set_title(f'{metric.capitalize()} across Folds', fontsize=12, weight='bold')
        axes[i].set_xlabel('Fold')
        axes[i].set_ylabel(metric.capitalize())
        axes[i].set_xticks(x_pos)
        axes[i].set_xticklabels([f'Fold {j+1}' for j in range(5)])
        axes[i].grid(True, alpha=0.3)
        axes[i].legend()

        # Add value labels on bars
        for j, bar in enumerate(bars):
            height = bar.get_height()
            axes[i].text(bar.get_x() + bar.get_width()/2., height + 0.005,
                        f'{height:.3f}', ha='center', va='bottom', fontsize=9)

    plt.suptitle('MKI_Rotation 5-Fold Cross Validation Results', fontsize=16, weight='bold')
    plt.tight_layout()
    plt.show()

# ---------------- Data Scaling Function ----------------
def scale_data(X_train, X_val):
    """Apply standard scaling to the data"""
    scaler = StandardScaler()

    # Reshape for scaling: (N, C, L) -> (N*L, C)
    n_train, c_train, l_train = X_train.shape
    X_train_reshaped = X_train.transpose(0, 2, 1).reshape(-1, c_train)
    scaler.fit(X_train_reshaped)
    X_train_scaled_reshaped = scaler.transform(X_train_reshaped)
    X_train_scaled = X_train_scaled_reshaped.reshape(n_train, l_train, c_train).transpose(0, 2, 1)

    n_val, c_val, l_val = X_val.shape
    X_val_reshaped = X_val.transpose(0, 2, 1).reshape(-1, c_val)
    X_val_scaled_reshaped = scaler.transform(X_val_reshaped)
    X_val_scaled = X_val_scaled_reshaped.reshape(n_val, l_val, c_val).transpose(0, 2, 1)

    return X_train_scaled, X_val_scaled

# ---------------- Main Execution ----------------
if __name__ == '__main__':
    # Configuration
    set_seed(42)

    # Dataset parameters
    PAMAP2_DATASET_PATH = '/content/'
    EXCLUDE_LABELS_PAMAP2 = [0]
    WINDOW_SIZE = 100
    STEP = 50

    # Training parameters
    BATCH_SIZE = 32  # Reduced from 32 to prevent memory issues
    EPOCHS = 100
    LEARNING_RATE = 0.001  # Reduced learning rate to prevent overfitting
    PATIENCE = 10  # Reduced patience for earlier stopping

    # Model parameters
    FEATURE_DIM = 32  # Reduced model complexity
    DROPOUT = 0.1  # Increased dropout for better regularization

    # Self-supervised parameters
    LABEL_RATIO = 0.2
    ALPHA = 1.0  # Reduced rotation loss weight

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # CUDA optimization settings
    if torch.cuda.is_available():
        print(f"CUDA Device: {torch.cuda.get_device_name()}")
        print(f"CUDA Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.2f} GB")
        torch.backends.cudnn.benchmark = True  # Optimize for fixed input size
        torch.cuda.empty_cache()  # Clear any cached memory

    # Load and preprocess data
    print("Loading and preprocessing PAMAP2 data...")
    X_windowed, y_encoded, CLASS_NAMES, NUM_CLASSES = load_and_window_pamap2(
        dataset_dir=PAMAP2_DATASET_PATH,
        window_size=WINDOW_SIZE,
        step=STEP,
        exclude_labels=EXCLUDE_LABELS_PAMAP2
    )

    if X_windowed is None:
        print("Data loading failed. Exiting.")
        exit()

    INPUT_CHANNELS = X_windowed.shape[1]
    print(f"Data shape: {X_windowed.shape}")
    print(f"Input channels: {INPUT_CHANNELS}, Number of classes: {NUM_CLASSES}")

    # Display class distribution
    class_counts = np.bincount(y_encoded)
    print(f"\nOverall class distribution:")
    for i, count in enumerate(class_counts):
        if i < len(CLASS_NAMES):
            print(f"  Class {i} ({CLASS_NAMES[i]}): {count} samples ({count/len(y_encoded)*100:.1f}%)")
        else:
            print(f"  Class {i}: {count} samples ({count/len(y_encoded)*100:.1f}%)")

    # 5-Fold Cross Validation
    print("\n" + "="*80)
    print("PAMAP2 MKI_Rotation Model - 5-Fold Cross Validation")
    print("="*80)

    kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    cv_results = {}
    all_confusion_matrices = []

    start_time = time.time()

    for fold_idx, (train_idx, val_idx) in enumerate(kfold.split(X_windowed, y_encoded)):
        print(f"\n{'='*50}")
        print(f"Training Fold {fold_idx + 1}/5")
        print(f"{'='*50}")

        # Split data for current fold
        X_train_fold = X_windowed[train_idx]
        X_val_fold = X_windowed[val_idx]
        y_train_fold = y_encoded[train_idx]
        y_val_fold = y_encoded[val_idx]

        # Check class distribution in this fold
        train_class_counts = np.bincount(y_train_fold)
        val_class_counts = np.bincount(y_val_fold)
        print(f"Fold {fold_idx+1} class distribution:")
        print(f"  Train: {train_class_counts}")
        print(f"  Val:   {val_class_counts}")

        # Scale data
        X_train_scaled, X_val_scaled = scale_data(X_train_fold, X_val_fold)

        print(f"Fold {fold_idx+1} - Train: {X_train_scaled.shape}, Val: {X_val_scaled.shape}")

        # Create datasets
        train_dataset = RotationDataset(X_train_scaled, y_train_fold, label_ratio=LABEL_RATIO)
        val_dataset = RotationDataset(X_val_scaled, y_val_fold, label_ratio=LABEL_RATIO)  # Same ratio for fair comparison

        # Create data loaders - Use single process to avoid multiprocessing issues in Colab
        num_workers = 0  # Disable multiprocessing for stability
        pin_memory = False  # Disable pin_memory to reduce memory pressure

        train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                                num_workers=num_workers, pin_memory=pin_memory)
        val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False,
                              num_workers=num_workers, pin_memory=pin_memory)

        # Initialize model
        model = Model_MKI_Rotation(INPUT_CHANNELS, FEATURE_DIM, NUM_CLASSES, DROPOUT)

        # Train model
        trained_model = train_self_supervised_model(
            model, train_loader, val_loader, device,
            epochs=EPOCHS, patience=PATIENCE, alpha=ALPHA, lr=LEARNING_RATE
        )

        # Evaluate model
        results = evaluate_model(trained_model, val_loader, device, CLASS_NAMES, fold_idx)
        cv_results[f'fold_{fold_idx+1}'] = results

        # Plot confusion matrix for this fold
        plot_confusion_matrix(results['labels'], results['predictions'], CLASS_NAMES, fold_idx)

        # Store confusion matrix
        cm = confusion_matrix(results['labels'], results['predictions'])
        all_confusion_matrices.append(cm)

        # Clean up memory more thoroughly to prevent issues
        del model, trained_model, train_dataset, val_dataset, train_loader, val_loader
        del X_train_scaled, X_val_scaled, X_train_fold, X_val_fold
        del y_train_fold, y_val_fold

        # Force garbage collection
        gc.collect()

        if torch.cuda.is_available():
            torch.cuda.empty_cache()
            torch.cuda.synchronize()  # Wait for all operations to complete

    end_time = time.time()
    total_time = end_time - start_time

    # ============================================================================
    # FINAL RESULTS SUMMARY
    # ============================================================================
    print("\n" + "="*80)
    print("FINAL CROSS-VALIDATION RESULTS")
    print("="*80)

    # Calculate mean and std for each metric
    metrics = ['accuracy', 'f1', 'precision', 'recall']
    final_results = {}

    for metric in metrics:
        values = [cv_results[f'fold_{i+1}'][metric] for i in range(5)]
        mean_val = np.mean(values)
        std_val = np.std(values)
        final_results[metric] = {'mean': mean_val, 'std': std_val, 'values': values}

    # Print detailed results for each fold
    print("\nDetailed Results by Fold:")
    print(f"{'Fold':<8} {'Accuracy':<10} {'F1 Score':<10} {'Precision':<10} {'Recall':<10}")
    print("-" * 60)
    for i in range(5):
        fold_key = f'fold_{i+1}'
        results = cv_results[fold_key]
        print(f"Fold {i+1:<3} {results['accuracy']:<10.4f} {results['f1']:<10.4f} {results['precision']:<10.4f} {results['recall']:<10.4f}")

    print("\nSummary Statistics:")
    print(f"{'Metric':<12} {'Mean':<10} {'Std':<10} {'Min':<10} {'Max':<10}")
    print("-" * 60)
    for metric in metrics:
        values = final_results[metric]['values']
        mean_val = final_results[metric]['mean']
        std_val = final_results[metric]['std']
        min_val = min(values)
        max_val = max(values)
        print(f"{metric.capitalize():<12} {mean_val:<10.4f} {std_val:<10.4f} {min_val:<10.4f} {max_val:<10.4f}")

    # Plot cross-validation results
    plot_cv_results(cv_results)

    # Create final results DataFrame
    print("\n" + "="*80)
    print("FINAL PERFORMANCE DATAFRAME")
    print("="*80)

    # Create individual fold results DataFrame
    fold_data = []
    for i in range(5):
        fold_key = f'fold_{i+1}'
        results = cv_results[fold_key]
        fold_data.append({
            'Fold': f'Fold {i+1}',
            'Accuracy': results['accuracy'],
            'F1_Score': results['f1'],
            'Precision': results['precision'],
            'Recall': results['recall']
        })

    # Add summary statistics
    fold_data.append({
        'Fold': 'Mean',
        'Accuracy': final_results['accuracy']['mean'],
        'F1_Score': final_results['f1']['mean'],
        'Precision': final_results['precision']['mean'],
        'Recall': final_results['recall']['mean']
    })

    fold_data.append({
        'Fold': 'Std',
        'Accuracy': final_results['accuracy']['std'],
        'F1_Score': final_results['f1']['std'],
        'Precision': final_results['precision']['std'],
        'Recall': final_results['recall']['std']
    })

    # Create and display DataFrame
    results_df = pd.DataFrame(fold_data)
    results_df = results_df.round(4)

    print("MKI_Rotation Model - 5-Fold Cross Validation Results:")
    print(results_df.to_string(index=False))

    # Additional summary
    print(f"\nTotal Training Time: {total_time:.2f} seconds ({total_time/60:.2f} minutes)")
    print(f"Average Time per Fold: {total_time/5:.2f} seconds")

    print(f"\nBest Fold Performance:")
    best_fold_idx = np.argmax([cv_results[f'fold_{i+1}']['f1'] for i in range(5)])
    best_fold_results = cv_results[f'fold_{best_fold_idx+1}']
    print(f"  Fold {best_fold_idx+1}: F1={best_fold_results['f1']:.4f}, Accuracy={best_fold_results['accuracy']:.4f}")

    print(f"\nFinal Model Performance (Mean  Std):")
    print(f"  Accuracy: {final_results['accuracy']['mean']:.4f}  {final_results['accuracy']['std']:.4f}")
    print(f"  F1 Score: {final_results['f1']['mean']:.4f}  {final_results['f1']['std']:.4f}")
    print(f"  Precision: {final_results['precision']['mean']:.4f}  {final_results['precision']['std']:.4f}")
    print(f"  Recall: {final_results['recall']['mean']:.4f}  {final_results['recall']['std']:.4f}")

    print("\n" + "="*80)
    print("Cross-validation completed successfully!")
    print("="*80)
```
